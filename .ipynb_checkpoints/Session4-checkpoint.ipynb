Import

```python
from bs4 import BeautifulSoup
import requests
import pandas
```

Read HTML

```python
with open ('IsoCode.txt','r') as fp:
    html = fp.read()
    soup = BeautifulSoup(html,'html5lib')
trs = soup.find_all('tr')
len(trs)
countries=[]

for tr in trs:
    values = [td.string for td in tr.find_all('td')]
    countries.append((str(values[0]),str(values[3])))
countries = countries[1:-1]
```

```python
import pickle
with open ('countries.pickle','wb+') as fp:
    pickle.dump(countries,fp)
```

Request Worldbank

```python
indicator_list={
'IMF WB\nCustoms and other import duties (% of tax revenue) 2016':'GC.TAX.IMPT.ZS',
'WB IMF Non-performing loans to total gross loans ratio (%)':'FB.AST.NPER.ZS',
'UIC Rail freight (mn ton x km travelled)':'IS.RRS.GOOD.MT.K6',
'WB Household final consumption expenditure, PPP bn USD':'NE.CON.PRVT.PP.CD',
'WB UNESCO R&D Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WB UNESCO Researchers p. mn inhabitants':'SP.POP.SCIE.RD.P6',
'IFDI Net Inflows (% of GDP)':'BX.KLT.DINV.WD.GD.ZS',
'WHO Current Health Expenditure p.c., PPP USD':'SH.XPD.CHEX.PP.CD',
'WB Research Expenditure % of GDP':'GB.XPD.RSDV.GD.ZS',
'WIPO Trademark applications':'IP.TMK.RSCT',
'WIPO Patent Applications':'IP.PAT.RESD'
}
```

```python
def world_bank(code,indicator):
    url = 'http://api.worldbank.org/v2/country/'+code+'/indicator/'+indicator+'?format=json'
    r = requests.get(url)
    try:
        data = r.json()
        for i in data [1]:
            if i['value'] is None:
                continue
            else :
                return i['value']
    except KeyError:
        return None
    except IndexError:
        return None
    except :
        return None
```

```python
df_v2=pandas.read_excel('/Users/dorianjodelet/Desktop/UTSEUS/Data_Science/Data-sheet-STUDENTS.xlsx',header=2,usecols='B:AR')

list = ['Country (according to CN Gov):']
for each in indicator_list :
    indicator_code = indicator_list[each]
    list.append(each)
    counter = 0
    print(each)
    
    for country_name, country_code in countries:
        df_v2.at[counter, 'Country (according to CN Gov):'] = country_name
        value = world_bank(country_code, indicator_code)
        df_v2.at[counter, each] = value
        if value is not None:
            df_v2.at[counter, each] = value
        counter+= 1
        #print(country_name,'done...')
        
df_v2
```
